```python
# import the required libraries
```

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
# Loading the dataset
```

```python
sonar_df = pd.read_csv("sonar_data.csv", header = None) # since header for data is not present in csv file
```

```python
# Preparing the data set
```

```python
X = sonar_df.drop(columns = 60, axis = 1).values
print("X")
print(X)
```

```python
y = sonar_df[60].values
print("y")
print(y)
```

```python
# Spliting the daset
```

```python
# stratify to keep proportion of R and M ie classes in target variable are maintained in both training and testing dataset
# stratify = output_label_name e.g y
# random_state is to shuffle for 1st time only and from next time onwards no shuffle of datasets
X_train, X_test , y_train, y_test = train_test_split(X, y, test_size = 0.1, stratify = y, random_state = 1)
```

```python
print("X_train")
print(X_train)
print()
print("y_train")
print(y_train)

print()
print("X_test")
print(X_test)
print()
print("y_test")
print(y_test)
```

```python
# LogisticRegression model since Binary classification problem
# M-> Mine
# R-> Rock
```

```python
# Model creation
```

```python
model = LogisticRegression()
```

```python
# Model training
```

```python
model.fit(X_train, y_train)
```

```python
# Model evaluation on training score data
```

```python
# accuracy_score(actual, predicted)
training_accuracy_score = accuracy_score(y_train, model.predict(X_train))
print("training_accuracy_score: ",training_accuracy_score)
```

```python
# Model evaluation on testing score data
```

```python
testing_accuracy_score = accuracy_score(y_test, model.predict(X_test))
print("testing_accuracy_score: ",testing_accuracy_score)
```
